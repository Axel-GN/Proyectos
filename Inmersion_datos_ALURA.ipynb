```python
# agrupar los datos por barrio y calcular la media del precio y la media del área
datos_barrio = inmuebles.groupby("Barrio").agg({"Precio_Millon": "mean", "Area": "mean"})

# calcular el precio por metro cuadrado de cada barrio
datos_barrio["Precio_m2"] = datos_barrio["Precio_Millon"] / datos_barrio["Area"]

# mostrar los datos por barrio ordenados de manera descendente por el precio por metro cuadrado
datos_barrio = datos_barrio.sort_values(by="Precio_m2", ascending=False)
print(datos_barrio)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# crear el gráfico de barras
top = datos_barrio.head(15)
plt.figure(figsize=(12,6))
sns.barplot(x=top.index, y="Precio_m2", data=top )

# personalizar el gráfico

plt.xticks(rotation=90) # rotar las etiquetas del eje x para mayor legibilidad
plt.ylabel("Precio por metro cuadrado") # etiquetar el eje y
plt.xlabel("Barrio") # etiquetar el eje x
plt.title("Top 15 de barrios por costo metro cuadrado") # agregar un título al gráfico
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
# mostrar el gráfico
#plt.savefig('/content/drive/MyDrive/valor_m2.png',format = 'png')
plt.show()

```
